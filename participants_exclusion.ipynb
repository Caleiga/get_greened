```python
import pandas as pd
```

```python
data = pd.read_csv('data_cleaned.csv', delimiter=",")
data = data.drop(columns=["Unnamed: 0", "Q25_First Click"])
```

```python
data.head(34)
```

```python
# remove veggie and vegan participants
index_veg = data.index[(data.diet == 1) | (data.diet == 2)]
excluded_participants = data.loc[index_veg]
excluded_participants["reason_to_exclude"] = "Veggie/Vegan"
data = data.drop(index_veg)
```

```python
# remove participants who didn't take any sandwich
index_noSandwich = data.index[data.sandwich == 5]
excluded_participants = pd.concat([excluded_participants, data.loc[index_noSandwich]])
excluded_participants["reason_to_exclude"] = excluded_participants["reason_to_exclude"].fillna("No sandwich taken")
data = data.drop(index_noSandwich)
```

```python
# Have a look at other diets
not_omni = data.drop(data.index[data.diet == 3])
not_omni.head(34) # we have 5 people, all said "other"
```

### Diets of participants that said "other"
55: no read meat
129: pesco-vegetarian
60: no particular diet
110: flexitarian (meat sandwich): Dr. Psycho, he told us his rule was only eat meat outside the house
120: Flexitarian: Vegetarian and Dairy free majority of the time; but when travelling or a guest am flexible with diet
-> I would say we keep 110 since his rule did allow him to eat meat and exclude 120 since their definition of flexitarian probably means that they would not eat meat in this situation

```python
# remove participant 55, 129, 120
index_otherDiet = data.index[(data.participant_number == 55) | (data.participant_number == 129) | (data.participant_number == 120)]
excluded_participants = pd.concat([excluded_participants, data.loc[index_otherDiet]])
excluded_participants["reason_to_exclude"] = excluded_participants["reason_to_exclude"].fillna("Other diets similar to veggie")
data = data.drop(index_otherDiet)
```

```python
# we remove participant 88 because he took the break at the wrong time
index_wrongBreak = data.index[data.participant_number == 88]
excluded_participants = pd.concat([excluded_participants, data.loc[index_wrongBreak]])
excluded_participants["reason_to_exclude"] = excluded_participants["reason_to_exclude"].fillna("Took the break at the wrong time")
data = data.drop(index_wrongBreak)
```

```python
excluded_participants = excluded_participants[["participant_number", "reason_to_exclude"]]
```
